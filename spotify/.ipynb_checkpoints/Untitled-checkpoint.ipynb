```python
# 日本のSpotifyユーザーが最も聴いている曲を取得する
top_tracks = spotify.playlist_tracks('37i9dQZEVXbKXQ4mDTEBXq', market='JP', limit=100)

# トップ100曲を表示
for index, track in enumerate(top_tracks['items']):
    track_name = track['track']['name']
    artist_name = track['track']['artists'][0]['name']
    print(f'{index + 1}. "{track_name}" by {artist_name}')
```

```python
artist_id = "4QvgGvpgzgyUOo8Yp8LDm9"
results = spotify.artist_albums(artist_id, album_type='single', country='JP', limit=20)

# ほしい情報を配列で取得する
artist_songs = []
for song in results['items'][:len(results)]:
    data = [
        '曲名：'+ song['name'], 
        '発売日：'+ song['release_date'], 
        'id：'+ song['id']]
    artist_songs.append(data)
pprint.pprint(artist_songs)
```
